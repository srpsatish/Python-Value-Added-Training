```python
#create a dtaframe from the following dict1={'a':1,'b':2,'c':3,'d':4}  dict2={'a':7,'b':8,'c':6,'d':9}
dict1={'a':1,'b':2,'c':3,'d':4}
dict2={'a':7,'b':8,'c':6,'d':9}
data={'first':dict1, 'second':dict2}
df=pandas.DataFrame(data)
df
```

```python
#data frame from 3 series
s1=pandas.Series([1,2,3,4,5,2,9])
s2=pandas.Series([1,2,3,4,5,2,9,5,4,9,6,7])
s3=pandas.Series(['a','b','c','d','e'])
data={'first':s1,'second':s2,'third':s3}
df=pandas.DataFrame(data)
df
```
